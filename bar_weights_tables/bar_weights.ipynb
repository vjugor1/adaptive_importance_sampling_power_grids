```python
import os
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
folders = {}
for fol in os.listdir():
    if 'pi' in fol and '.' not in fol:
        files = os.listdir(fol)
        folders[fol] = [os.path.join(fol, file) for file in files]
```

# $\pi/ 2$

```python
#transparency and contour 
plt.rcParams['font.size'] = '14'
for json_file in folders['pi2']:
    plt.figure(figsize=(10, 5))
    #plt.title(r'$\max \theta_{ij} \leq \pi / 2$ ' + json_file.split('\\')[1].split('.')[0])
    with open(json_file, 'r') as fp:
        data = json.load(fp)
    plt.bar(range(len(data['MD-Var-W'])), np.sort(data['MD-Var-W'])[::-1], alpha = 0.6, label='MD-Var')
    #plt.bar(range(len(data['MD-KL-W'])), np.sort(data['MD-KL-W'])[::-1], alpha = 0.6, label='MD-KL')
    plt.bar(range(len(data['ALOE-W'])), np.sort(data['ALOE-W'])[::-1], alpha = 0.6, label='ALOE')
    plt.xlabel("Number of Plane")
    plt.ylabel("Weight Value")
    plt.legend()
    name_grid = json_file.split('\\')[1].split('.')[0]
    if name_grid == 'grid118i' or name_grid == 'grid3120':
        plt.savefig("weights_pi2_" + name_grid + ".jpg")
    plt.show()
```
