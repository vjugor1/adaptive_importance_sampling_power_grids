```python
import os
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {}
with open('hist_data.json', 'r') as fp:
    data = json.load(fp)
with open('hist_data.txt', 'r') as fp:
    data['pmvnorm'] = np.loadtxt(fp)
```

```python
plt.rcParams['font.size'] = '24'
plt.figure(figsize=(15, 7))
for key in data.keys():
    if key != 'MC':
        if key == 'pmvnorm':
            plt.hist(data[key], label=key, alpha=0.6, rwidth=0.8, bins=150, edgecolor='black')
        else:
            plt.hist(data[key], label=key, alpha=0.6, rwidth=0.8, edgecolor='black')

plt.xlim((0.48, 1.4))
plt.legend()
#plt.grid()
plt.xlabel('Relative Estimate')
plt.ylabel('Frequency')
plt.savefig('histogram.jpg')
plt.show()
```
