```python
import os
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
with open('dict_hist_shiv.txt', 'r') as fp:
    data_ = fp.readlines()
```

```python
name_map = {}
name_map['MD'] = 'MD-Var'
name_map['OMC'] = 'ALOE'
```

```python
data = {}
for guy in data_:
    if guy.split('\t')[0] == 'OMC' or guy.split('\t')[0] == 'MD':
        print(guy.split('\t')[0])
        data[name_map[guy.split('\t')[0]]] = np.array(guy.split('\t')[1][1:-2].split(', '), dtype=np.float)
```

```python
plt.rcParams['font.size'] = '24'
plt.figure(figsize=(15, 7))
for key in data.keys():
    if key == 'ALOE':
    plt.hist(np.array(data[key]), label=key, alpha=0.6, rwidth=0.8, bins=10, edgecolor='black')

#plt.xlim((0.48, 1.4))
#plt.xticks([min(data['OMC']), (min(data['OMC']) + max(data['OMC'])) / 2, max(data['OMC'])])
plt.legend()
#plt.grid()
plt.xlabel('Relative Estimate')
plt.ylabel('Frequency')
plt.savefig('histogram.jpg')
plt.show()
```
