```python
import os
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
json_files = []
for file in os.listdir():
    if '.json' in file:
        json_files.append(file)
```

```python
#118, 1354
plt.rcParams['font.size'] = '14'
for json_file in json_files:
    plt.figure(figsize=(10, 5))
    #plt.title('Distribution of Simultaneous Number of Planes Violation\n' + json_file.split('.')[0])
    with open(json_file, 'r') as fp:
        data = json.load(fp)
    xs = []
    ys = []
    for key in data.keys():
        if 'Ineqs.' in key:
            break
        xs.append(int(key))
        ys.append(data[key])
    plt.bar(xs, ys)
    plt.xlabel('Number of Simultaneous Plane Violations')
    plt.ylabel('Frequency')
    #plt.bar(range(len(data['MD-Var-W'])), np.sort(data['MD-Var-W'])[::-1], alpha = 0.6, label='MD-Var')
    #plt.bar(range(len(data['MD-KL-W'])), np.sort(data['MD-KL-W'])[::-1], alpha = 0.6, label='MD-KL')
    #plt.bar(range(len(data['ALOE-W'])), np.sort(data['ALOE-W'])[::-1], alpha = 0.6, label='ALOE')
    #plt.legend()
    plt.savefig('Distribution_of_Simultaneous_Number_of_Planes_Violation_' + json_file.split('.')[0] + '.jpg')
    plt.show()
    
```
